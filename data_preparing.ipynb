```python
# xử lý tuổi, thu nhập, năm làm việc về dạng chuẩn
# chuyển tuổi, năm làm việc về số dương
train_df_copy["Age"] = np.abs(train_df_copy["Age"])
train_df_copy["Employment length"] = np.abs(train_df_copy["Employment length" ])

# xử lý tuổi về hưu: employment length = 365243 được chuyển thành 0
train_df_copy["Employment length"][train_df_copy["Employment length"] == 365243] = 0
```

```python
from mlxtend.preprocessing import minmax_scaling
```

```python
train_df_copy["Age"] = minmax_scaling(train_df_copy["Age"], columns=[0])
train_df_copy["Employment length"] = minmax_scaling(train_df_copy["Employment length"], columns=[0])
train_df_copy["Income"] = minmax_scaling(train_df_copy["Income"], columns=[0])

train_df_copy
```

chuyển các giá trị của các cột, Gender, Has a car, has a property, Employment status, education level, marital status, dweling về dạng số

```python
# gender
for i in range(len(train_df_copy)):
    if train_df_copy["Gender"].iloc[i] == 'F':
        train_df_copy["Gender"].iloc[i] = 0
    else:
        train_df_copy["Gender"].iloc[i] = 1

train_df_copy
```

```python
def convertYNtoNum(df, feature):
    for i in range(len(df)):
        if df[feature].iloc[i] == 'Y':
            df[feature].iloc[i] = 1
        else:
            df[feature].iloc[i] = 0
```

```python
convertYNtoNum(train_df_copy, 'Has a car')
convertYNtoNum(train_df_copy, 'Has a property')
```
